```python
main = merged_data.loc[:, 'Население':'Численность других отделений']
main['M1 Вовлеченность'] = main['Общее количество вовлеченной молодежи']/main['Население']/25
main['Бюджет на человек'] = main['Бюджет']/(main['Общее количество вовлеченной молодежи']/25)
main['M2 Вовлеченность'] = 0.01/(main['Бюджет']/(main['Общее количество вовлеченной молодежи']/25)/ main['Бюджет на человек'].max())
main['M3 Итоговая'] = 2*main['M1 Вовлеченность']*main['M2 Вовлеченность']/(main['M1 Вовлеченность']+main['M2 Вовлеченность'])
M1q40 = main['M1 Вовлеченность'].quantile(0.4)
M1q80 = main['M1 Вовлеченность'].quantile(0.8)
def p_group(budget):   
    if budget <= M1q40:
        return 'Низкий'
    if budget <= M1q80:
        return 'Средний'
    return 'Высокий'
main['Уровень вовлеченности населения'] = main['M1 Вовлеченность'].apply(p_group)
M2q40 = main['Бюджет на человек'].quantile(0.01)
M2q80 = main['Бюджет на человек'].quantile(0.8)
def b_group(budget):   
    if budget <= M2q40:
        return 'Низкий'
    if budget <= M2q80:
        return 'Средний'
    return 'Высокий'
main['Уровень финансирования населения'] = main['M1 Вовлеченность'].apply(b_group)

#main
```
