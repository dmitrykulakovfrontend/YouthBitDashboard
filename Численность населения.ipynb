```python
import pandas as pd
```

```python
def population(area, num_of_obl, name_of_area):
    final = pd.DataFrame({'Область': [], 'Население': [], 'Субъект': []})
    for i in range(1, num_of_obl+1):
        sheet = '2.' + str(area) + '.' + str(i) + '.'
        df = pd.read_excel('Бюллетень_2021.xlsx', sheet_name=sheet)
        summ = df.loc[26, 'Unnamed: 1'] + df.loc[58, 'Unnamed: 1'] + df.loc[59, 'Unnamed: 1']
        obl = df.loc[0, 'ЧИСЛЕННОСТЬ НАСЕЛЕНИЯ ПО ПОЛУ И ВОЗРАСТУ  \n на 1 января 2021 года ']
        df2 = pd.DataFrame({'Область': [obl], 'Население': [summ], 'Субъект': [name_of_area]})
        final = final.append(df2, ignore_index = True )
    return final
```

```python
pop = pd.concat([population(1, 18, 'ЦФО'), \
                 population(2, 12, 'СЗФО'), \
                 population(3, 8, 'ЮФО'), \
                 population(4, 7, 'СКФО'), \
                 population(5, 14, 'ПФО'), \
                 population(6, 7, 'УФО'), \
                 population(7, 10, 'СФО'), \
                 population(8, 11, 'ДФО')], ignore_index=True)
```

```python
pop.to_excel('Молодое население.xlsx', index= False)
```
